### 1. What is name of car that has the highest horsepower?

### 2. How many cars have mpg ≥ 35?

```python
#your code here
count_cars = (df['mpg'] >= 35).sum()
print("Number of cars with mpg ≥ 35:", count_cars)
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
#your code here
filtered_cars = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]

most_common_origin = filtered_cars['origin'].mode()[0]

print("Most common origin for cars with horsepower > 100 and weight < 3000:", most_common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
japan_cars = df[df['origin'].str.strip().str.lower() == 'japan']
mean_acceleration_japan = japan_cars['acceleration'].mean()
print("Mean acceleration of cars from Japan:", round(mean_acceleration_japan, 2))
```

### 5. Which year had the highest average mpg?

```python

avg_mpg_by_year = df.groupby('model_year')['mpg'].mean()

year_with_highest_avg_mpg = avg_mpg_by_year.idxmax()
highest_avg_mpg = avg_mpg_by_year.max()

print(f"Year with highest average mpg: {year_with_highest_avg_mpg} (Average mpg: {highest_avg_mpg:.2f})")

```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python

median_mpg = df['mpg'].median()

cars_above_median_mpg = df[df['mpg'] > median_mpg]

cars_above_median_mpg['hp_to_weight'] = cars_above_median_mpg['horsepower'] / cars_above_median_mpg['weight']

max_ratio = cars_above_median_mpg['hp_to_weight'].max()

best_cars = cars_above_median_mpg[cars_above_median_mpg['hp_to_weight'] == max_ratio]

print(best_cars[['name', 'horsepower', 'weight', 'mpg', 'hp_to_weight']])

```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import matplotlib.pyplot as plt
import seaborn as sns

avg_mpg_origin_year = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_mpg_origin_year, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Evolution of Average MPG Over Years by Origin')
plt.xlabel('Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.show()

```

### Create a Seaborn scatterplot (or PairGrid) where:

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
import pandas as pd

df_reset = df.reset_index() if df.index.name == 'name' else df.copy()

grouped = df_reset.groupby('name').agg(
    num_years=('model_year', 'nunique'),       
    mpg_std=('mpg', 'std'),              
    mpg_mean=('mpg', 'mean')            
).reset_index()

consistent_cars = grouped[(grouped['num_years'] >= 2) & (grouped['mpg_std'] < 1.0)]

min_std = consistent_cars['mpg_std'].min()

lowest_variation_cars = consistent_cars[consistent_cars['mpg_std'] == min_std]

result = lowest_variation_cars.sort_values(by=['num_years', 'mpg_mean'], ascending=[False, False])

print(result[['name', 'num_years', 'mpg_mean']])

```
